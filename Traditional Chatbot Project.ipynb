```python
def chatbot():
    print("Hello! I am your assistant chatbot.")
    print("Type 'help' to see what I can do. Type 'exit' to quit.")

    while True:
        user_input = input("You: ").lower().strip()

        if user_input == "exit":
            print("Chatbot: Goodbye!")
            break

        elif user_input == "help":
            print("Chatbot: I can respond to the following:")
            print("- greetings (hi, hello)")
            print("- ask about name")
            print("- ask about capabilities")
            print("- simple questions")

        elif user_input in ["hi", "hello"]:
            print("Chatbot: Hello! How can I help you?")

        elif "name" in user_input:
            print("Chatbot: I am a simple rule-based chatbot.")

        elif "how are you" in user_input:
            print("Chatbot: I'm just code, but I'm running perfectly!")

        elif "what can you do" in user_input:
            print("Chatbot: I can answer simple questions and guide you.")

        elif user_input == "":
            print("Chatbot: Please enter something.")

        else:
            print("Chatbot: Sorry, I didn’t understand that. Type 'help'.")

chatbot()
```
